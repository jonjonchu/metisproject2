# Score Tester
Implement score that only tracks previous films as metric

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path=('./data/mojo_all_years_movies_cleaned.pkl')
df = pd.read_pickle(path)
```

```python
categories = ['Director', 'Domestic_Distributor','Cast1']
```

```python
for category in categories:
    score_col_name = category+'_Score'
    score_col_name

    df[score_col_name] = (df.sort_values(['Release_Date'])
                                .groupby([category])
                                .Domestic_Total_Gross
                                .apply(lambda x: x.expanding().mean().shift())
                            )
    # take log of score
    df[score_col_name] = df[score_col_name].apply(lambda x: np.log(x))
```
