# Crew parser

```python
from bs4 import BeautifulSoup
import requests
import re
import numpy as np
```

```python
url = 'https://www.boxofficemojo.com/title/tt3794354/credits/?ref_=bo_tt_tab'

response = requests.get(url)
print('requests.get status: ',response.status_code)
page = response.text
soup = BeautifulSoup(page, "lxml")
del page, response
```

```python
crewInfo = soup.find(id="principalCrew").find_all('tr')
for crew in crewInfo:
    print('+++++++++++++++++++++++++++++++++')
    print(crew.prettify())
```

```python
crew_list = []
director = None
writer = None
producer = None
cinematographer = None
```

```python
for crew in crewInfo[1:]:
    
    crew_member = crew.find('td').text.split('\n')[0]
    role = crew.find('td').find_next('td').text
    #print([role, crew_member])
    
    if director is None:
        if role == 'Director':
            director = crew_member
    if writer == None:
        if role == 'Writer':
            writer = crew_member
    if producer == None:
        if role == 'Producer':
            producer = crew_member
    if cinematographer == None:
        if role == 'Cinematographer':
            cinematographer = crew_member
```

```python
# Will just take first instance of Director, Writer, Producer, Cinematographer
print(director, writer, producer, cinematographer)
```
